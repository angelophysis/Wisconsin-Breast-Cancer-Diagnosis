This model attempts to predict whether a tumor is malignant or benign based on given medical characteristics.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
uri = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
dataset = pd.read_csv(uri, header=None)

dataset
```

The labels are the "diagnosis" column and the features are the other columns.

```python
x = dataset.drop(columns=[1])
y = dataset[1]
```

As the features have different scales, we need to standardize them before training the SVM model. For that, we use StandardScaler from scikit-learn.

```python
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

Now, we separate the training data and test data, train the model, make predictions and calculate accuracy.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=10, test_size=0.25, stratify=y)

model = SVC()
model.fit(x_train, y_train)

predictions = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, predictions)

print('Accuracy: {:.2f}%'.format(accuracy*100))
```
